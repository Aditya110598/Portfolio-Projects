```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load dataset

athletes = pd.read_csv('F:/Projects/Python/Olympics Data Analysis/athlete_events.csv')
regions = pd.read_csv('F:/Projects/Python/Olympics Data Analysis/noc_regions.csv')
```

```python
athletes.head()
```

```python
regions.head()
```

```python
# Join the dataframes using Merge function of pandas

athletes_df = athletes.merge(regions , how ='left', on = 'NOC')
athletes_df.head()
```

```python
athletes_df.shape
```

```python
# Changing Column Names

athletes_df.rename(columns = {'region' : 'Region', 'notes' : 'Notes'}, inplace = True)
```

```python
athletes_df.head()
```

```python
athletes_df.info()
```

```python
athletes_df.isnull().sum()
```

```python
# India Details

athletes_df.query('Team == "India"').head()
```

```python
# Japan Details

athletes_df.query('Team == "Japan"').head()
```

```python
# Top countries participating since inception of Olympics

top_10_countries = athletes_df.Team.value_counts().sort_values(ascending = False).head(10)
top_10_countries
```

```python
# Plot for top 10 countries

plt.figure(figsize=(12,6))
plt.title('Overall Participation by Country')
sns.barplot(x = top_10_countries.index, y = top_10_countries)
```

```python
# Age Distribution of the Participants

plt.figure(figsize=(12,6))
plt.title('Age distribution of the athletes')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age, bins = np.arange(10,80,2), color = 'green', edgecolor = 'white');
```

```python
# Winter Olympics Sports

winter_sports = athletes_df[athletes_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
# Summer Olympics Sports

Summer_sports = athletes_df[athletes_df.Season == 'Summer'].Sport.unique()
Summer_sports
```

```python
# Male and Female Participants

gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
# Pie plot for Male and female Participants

plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%',  shadow = True)
```

```python
# Total medals won by all athletes

athletes_df.Medal.value_counts()
```

```python
# Total number of female athletes taken part in each Summer olympics

female_participants = athletes_df[(athletes_df.Sex =='F') & (athletes_df.Season == 'Summer')][['Sex','Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.head()
```

```python
female_participants.tail()
```

```python
women_participants = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')]
women_participants.head()
```

```python
plt.figure(figsize = (20,10))
sns.countplot(x='Year', data =women_participants)
plt.title("Women Participation")
```

```python
# Total number of male athletes taken part in each Summer olympics

male_participants = athletes_df[(athletes_df.Sex =='M') & (athletes_df.Season == 'Summer')][['Sex','Year']]
male_participants = male_participants.groupby('Year').count().reset_index()
male_participants.head()
```

```python
male_participants.tail()
```

```python
# Gold medal athletes

gold_medalists = athletes_df[(athletes_df.Medal == 'Gold')]
gold_medalists
```

```python
# Gold medals from each country

gold_medalists.Region.value_counts().reset_index(name = 'Medal').head(5)
```

```python
total_gold_medals = gold_medalists.Region.value_counts().reset_index(name = 'Medal').head(6)
g = sns.catplot(x = 'index', y='Medal', data = total_gold_medals, height = 5, kind='bar')
g.set_xlabels('Top 5 Countries')
g.set_ylabels('Number of medals')
plt.title('Gold medals per country')
```

```python
# Rio Olympics

max_year = athletes_df.Year.max()
print(max_year)

team_names = athletes_df[(athletes_df.Year == max_year) & (athletes_df.Medal == 'Gold')].Team

team_names.value_counts().head(10)
```

```python
sns.barplot(x = team_names.value_counts().head(20), y = team_names.value_counts().head(20).index)
plt.ylabel(None)
plt.xlabel('Countrywise Medals for 2016 Rio Olympics')
```
