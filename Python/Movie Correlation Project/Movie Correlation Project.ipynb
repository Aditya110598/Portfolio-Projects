```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

The above code performs numerical encoding of categorical variables in a pandas DataFrame df.

1.creates a copy of the original DataFrame df and assigns it to a new DataFrame object df_numerized.

2.This loop iterates over each column in the DataFrame df_numerized and checks if the column's data type is object, which is pandas way of representing string or categorical data.

3.If the column's data type is object, the code first converts the column to a pandas category data type using the astype() method.

4.Then, it uses the cat.codes attribute of the pandas category type to convert each unique category in the column to a numerical code.

5.Finally, the code outputs the DataFrame df_numerized with the categorical variables replaced with their numerical codes. This DataFrame can be used for further analysis or modeling tasks that require numerical data.

```python
correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values()

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

High_corr = sorted_pairs[abs(sorted_pairs) > 0.5]

print(High_corr)
```
